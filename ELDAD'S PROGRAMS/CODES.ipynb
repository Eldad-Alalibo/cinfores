```python
#WRITE A PROGRAM THAT TAKES NUMBERS FROM THE USER UNTIL THEY SAY 'DONE'
#AND CALCULATE THE SUM,COUNT AND AVERAGE OF THE NUMBERS AND DISPLAY THEM

sum=0
count=0
while True:
    Num = input('enter a number or type "done" if you want to get the sum,average and count of the number(s)')
    if Num == 'done' or Num=='Done':
        print('That would be all')
        break
    else:
        count+=1
        Num=float(Num)
        sum= sum+ Num
        Average=sum/count
        
if count > 0:
    print('The sum is:',sum)
    print('The count is: ', count)
    print('The average is: ', Average)
    
    
        
    
```

```python
#write a program that takes a string and displays
# only the vowels
string = input ('enter a string')
print ('vowels:',end=' ')
for character in string:
    if character in 'aeiouAEIOU':
        print(character,end=' ')
```

```python
#string methods
string = input ('enter a string')
print ('consonants', end= ' ')
for character in string:
    if character in 'bcdfghjklmnpqrstvwxyxBCDFGHJKLMNPQRSTVWXYZ':
        print(character,end=' ')
```
